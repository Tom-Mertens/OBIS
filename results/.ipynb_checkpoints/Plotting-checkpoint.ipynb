```python
import seaborn as sns
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
pd.set_option('display.max_rows', 500)
import os


with open('OBIS_results.pickle', 'rb') as f:
    results = pickle.load(f)
```

```python
lof = results['T1']['North']['lof_df']
```

```python
def plot_heatmap(lof):
    fig, ax = plt.subplots(facecolor='w')
    sns.heatmap(lof.corr())
    plt.title("Correlations for T1 North")
    plt.xlabel('Intersection')
    plt.ylabel('Intersection')
    plt.show()
    return
```

```python
for trajectory in results.keys():
    for direction in results[trajectory].keys():
        print(trajectory,direction)
        lof = results[trajectory][direction]['lof_df']
        plot_heatmap(lof)
```

```python
def compare_flows(outlier, intersections, direction='North',trac='T1',results=results):
    """Compare flows for an outlier moment at different intersections"""
    outlier = pd.Timestamp(outlier)
    
    fig, ax = plt.subplots(figsize=(10,8), facecolor='w')
    xticks = [x for x in range(0,60,5)]
    
    #first establish baseline:
    base_df = results[trac][direction]['fpds'][intersections[0]]
    outlier_data = base_df.loc[(base_df['timestamp'].dt.dayofweek ==outlier.weekday()) & ((base_df['timestamp'].dt.hour==outlier.hour))]
    outlier_data= outlier_data['prob'].groupby(by=outlier_data['timestamp'].dt.time).agg([np.mean, np.std])

    plt.plot(xticks,outlier_data['mean'],label='Average FPD') #plot the average FPDs   
    plt.fill_between(xticks, outlier_data['mean']-outlier_data['std'],outlier_data['mean']
                     +outlier_data['std'], facecolor='#C1DBFF', alpha=1.0, edgecolor='none')
    
    for intersection in intersections:
        df = results[trac][direction]['fpds'][intersection]
        df = df.loc[df['timestamp']>=outlier][:12]
        
        plt.plot(xticks,df['prob'],label=intersection) #Plot second flow
    plt.grid()
    plt.legend(fontsize=18)
    plt.title('Outlier: 2019-01-01 07:00:00', fontsize = 20)
    plt.suptitle('Comparison of correlating and uncorrelating intersections', fontsize = 20)
    plt.xlabel('Minutes after 07:00')
    plt.ylabel('Traffic Proportion (FPD value)')
    plt.show()
    
    return outlier_data, base_df
a,b = compare_flows("2019-01-01 07:00:00",['K504','K263','K561'])
```
